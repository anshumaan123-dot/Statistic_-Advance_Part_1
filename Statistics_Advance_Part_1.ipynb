#Theory Questions

Q1. What is a random variable in probability theory?
- In probability theory, a random variable is a variable that represents the outcome of a random experiment. It is a function that assigns a numerical value to each possible outcome in the sample space of a probabilistic experiment.

Q2. What are the types of random variables?
- Random variables are primarily classified based on the type of values they can take and how those values are distributed. In probability theory, there are two main types of random variables, with some subtypes or variations. Here's a breakdown:
1. Discrete Random Variable
2. Continuous Random Variable
3. Mixed Random Variable
4. Categorical Random Variable (mainly in statistics/data science)

Q3. What is the difference between discrete and continuous distributions?
- A discrete distribution describes the probability of outcomes of a discrete random variable, which can take on a finite or countably infinite number of distinct values, such as the roll of a die or number of phone calls in an hour. In contrast, a continuous distribution represents a continuous random variable that can assume any value within a range or interval, like temperature or time. While discrete distributions use probability mass functions (PMFs), continuous distributions use probability density functions (PDFs). Probabilities in discrete cases are assigned to exact values, while in continuous cases, they are calculated over intervals.

Q4. What are probability distribution functions (PDF)?
- A Probability Distribution Function (PDF) is a function that describes the likelihood of a continuous random variable taking on a particular value. Since the probability of a continuous variable being exactly any single value is zero, the PDF instead gives the density of the probability around that value.

Q5. How do cumulative distribution functions (CDF) differ from probability distribution functions (PDF)?
- A Cumulative Distribution Function (CDF) gives the probability that a random variable is less than or equal to a specific value, while a Probability Distribution Function (PDF) shows the relative likelihood of the variable taking on a particular value. For continuous variables, the CDF is the integral of the PDF, providing cumulative probabilities, whereas the PDF represents probability density, not exact probabilities.

Q6. What is a discrete uniform distribution?
- A Discrete Uniform Distribution is a probability distribution where each outcome in a finite set has an equal probability of occurring.

Q8. What is the binomial distribution, and how is it used in probability?

Q9. What is the Poisson distribution and where is it applied?

Q10. What is a continuous uniform distribution?
- The continuous uniform distribution is a type of continuous probability distribution in which all values within a certain interval are equally likely to occur. It is the simplest form of a continuous distribution.

Q11. What are the characteristics of a normal distribution?
- The normal distribution, also known as the Gaussian distribution, is one of the most important probability distributions in statistics. It describes how values of a continuous random variable are distributed in many natural and social phenomena.

Q12. What is the standard normal distribution, and why is it important?
- The standard normal distribution is a special case of the normal distribution where the mean (μ) is 0 and the standard deviation (σ) is 1.
- It is Important because:
1. Simplifies Calculations : Complex normal distributions can be converted to the standard normal.
2. Enables Use of Z-Tables : Probabilities and percentiles for any normal distribution can be found using Z-tables, which are based on the standard normal curve.
3. Universal Benchmark : It provides a common reference for comparing different datasets or distributions.
4. Foundational for Statistical Inference : Used in hypothesis testing, confidence intervals, control charts, and many other statistical methods.

Q13. What is the Central Limit Theorem (CLT), and why is it critical in statistics?
- The Central Limit Theorem (CLT) is a fundamental concept in statistics that states : " When independent random variables are added together, their normalized sum tends toward a normal distribution — even if the original variables themselves are not normally distributed — as the sample size becomes large."
- It is Critical in statistics because:
1. Enables Use of Normal Distribution : Allows statisticians to apply normal distribution techniques (z-scores, confidence intervals, hypothesis tests) to problems involving means, even if the underlying data is not normal.
2. Foundation for Inferential Statistics : Justifies using sample data to make inferences about population parameters.
3. Applies Broadly : Works for any distribution (skewed, discrete, etc.) as long as sample size is large and data meets basic conditions (independent and identically distributed).
4. Simplifies Analysis : Makes analysis practical and powerful, especially when working with large datasets or unknown population distributions.

Q14. How does the Central Limit Theorem relate to the normal distribution?
- The Central Limit Theorem (CLT) relates directly to the normal distribution by explaining why and when the normal distribution appears in statistics, even when the data itself is not normally distributed.

Q15. What is the application of Z statistics in hypothesis testing?
- Z-statistics (or Z-scores) play a crucial role in hypothesis testing, especially when dealing with large sample sizes or known population parameters. They help determine how far a sample statistic (like the sample mean) is from the null hypothesis, measured in units of standard deviation.

Q16.  How do you calculate a Z-score, and what does it represent?
- A Z-score (also called a standard score) tells you how many standard deviations a data point is from the mean of a distribution. It allows you to compare values from different distributions or assess how unusual a value is.
- It Represent:
1. Z=0 : Value is exactly at the mean
2. Z = +1 : Value is 1 standard deviation above the mean
3. Z = -2 :  Value is 2 standard deviations below the mean

Q17. What are point estimates and interval estimates in statistics?
- In statistics, both point estimates and interval estimates are methods used to estimate population parameters (like mean, proportion, or standard deviation) based on sample data. However, they differ in how they express uncertainty.

Q18. What is the significance of confidence intervals in statistical analysis?
- Confidence intervals (CIs) are essential in statistical analysis because they provide a range of plausible values for a population parameter (like mean or proportion) based on sample data — along with an associated confidence level (e.g., 95%).

Q19.  What is the relationship between a Z-score and a confidence interval?
- The Z-score and confidence interval are closely related in statistics, especially when estimating population parameters using normal distributions. The Z-score helps define the range of a confidence interval.

Q20.  How are Z-scores used to compare different distributions?
- Z-scores are a powerful tool for comparing values from different distributions, even when those distributions have different means and standard deviations. They standardize values by converting them into a common scale — the standard normal distribution.

Q21. What are the assumptions for applying the Central Limit Theorem?
- The Central Limit Theorem (CLT) is a powerful tool in statistics, but it relies on certain assumptions to be valid. These assumptions ensure that the sampling distribution of the sample mean approximates a normal distribution, even if the population is not normally distributed.
- Assumptions for Applying the Central Limit Theorem:
1. Independent Observations
2. Identically Distributed
3. Sample Size Should Be Sufficiently Large
4. Finite Variance and Mean

```python
#3.Write a Python function to calculate the probability distribution function (PDF) of a Bernoulli distribution.

def bernoulli_pdf(x, p):
    if x not in [0, 1]:
        return 0.0  # PDF is 0 for invalid outcomes
    return p if x == 1 else 1 - p

# Define probability of success
p = 0.6

# Calculate and print probabilities
print(f"P(X=0): {bernoulli_pdf(0, p)}")  # Expected: 0.4
print(f"P(X=1): {bernoulli_pdf(1, p)}")  # Expected: 0.6
print(f"P(X=2): {bernoulli_pdf(2, p)}")  # Expected: 0.0 (invalid)
```

```python
#9.Write a Python function to calculate Z-scores from a dataset and plot them.

import numpy as np
import matplotlib.pyplot as plt

def calculate_and_plot_z_scores(data):
    data = np.array(data)
    mean = np.mean(data)
    std_dev = np.std(data)

    # Calculate Z-scores
    z_scores = (data - mean) / std_dev

    # Plot original data and Z-scores
    plt.figure(figsize=(10, 5))

    # Plot original data
    plt.subplot(1, 2, 1)
    plt.plot(data, 'bo-', label='Original Data')
    plt.title('Original Data')
    plt.xlabel('Index')
    plt.ylabel('Value')
    plt.grid(True)
    plt.legend()

    # Plot Z-scores
    plt.subplot(1, 2, 2)
    plt.plot(z_scores, 'ro-', label='Z-scores')
    plt.axhline(0, color='black', linestyle='--')
    plt.title('Z-scores')
    plt.xlabel('Index')
    plt.ylabel('Z-score')
    plt.grid(True)
    plt.legend()
    plt.tight_layout()
    plt.show()
    return z_scores

# Example dataset
data = [10, 12, 9, 15, 14, 13, 10, 8, 11, 16]

# Call the function
z_scores = calculate_and_plot_z_scores(data)

# Print the Z-scores
print("Z-scores:", z_scores)
```

```python
#14.Write a Python program to calculate the Z-score for a given data point and compare it to a standard normal distribution.

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def calculate_z_score(x, mean, std_dev):
    return (x - mean) / std_dev

def plot_standard_normal_with_z(z_score):
    x = np.linspace(-4, 4, 1000)
    y = norm.pdf(x, 0, 1)

    plt.figure(figsize=(10, 5))
    plt.plot(x, y, label='Standard Normal PDF', color='blue')
    plt.fill_between(x, 0, y, where=(x <= z_score), color='red', alpha=0.4, label=f'Area ≤ Z={z_score:.2f}')
    plt.axvline(z_score, color='red', linestyle='--', label=f'Z-score = {z_score:.2f}')

    plt.title('Standard Normal Distribution and Z-score')
    plt.xlabel('Z-score')
    plt.ylabel('Probability Density')
    plt.grid(True, linestyle='--', alpha=0.6)
    plt.legend()
    plt.show()

# Example: Data point
x = 85
mean = 70
std_dev = 10

# Calculate Z-score
z = calculate_z_score(x, mean, std_dev)
print(f"Z-score for x = {x} (mean = {mean}, std_dev = {std_dev}): {z:.2f}")

# Visualize
plot_standard_normal_with_z(z)
```

```python
#15.Implement hypothesis testing using Z-statistics for a sample dataset.

import numpy as np
from scipy.stats import norm

def z_test(sample_data, population_mean, population_std, alpha=0.05):
    n = len(sample_data)
    sample_mean = np.mean(sample_data)

    # Calculate Z-statistic
    z_score = (sample_mean - population_mean) / (population_std / np.sqrt(n))

    # Two-tailed p-value
    p_value = 2 * (1 - norm.cdf(abs(z_score)))

    # Decision
    reject_null = p_value < alpha

    result = {
        'sample_mean': sample_mean,
        'z_score': z_score,
        'p_value': p_value,
        'reject_null': reject_null,
        'conclusion': 'Reject H₀' if reject_null else 'Fail to reject H₀'
    }
    return result

# Example: Sample data
sample_data = [52, 49, 51, 48, 50, 53, 47, 49, 52, 50]
population_mean = 50
population_std = 2  # known population std deviation

# Run Z-test
results = z_test(sample_data, population_mean, population_std)

# Print results
print("Z-Test Results:")
for key, value in results.items():
    print(f"{key}: {value}")
```

```python
#16.Create a confidence interval for a dataset using Python and interpret the result.

import numpy as np
from scipy import stats

def confidence_interval(data, confidence=0.95):
    data = np.array(data)
    n = len(data)
    mean = np.mean(data)
    sem = stats.sem(data)  # Standard error of the mean

    # Calculate margin of error using the t-distribution
    margin = sem * stats.t.ppf((1 + confidence) / 2, df=n-1)
    return mean - margin, mean + margin

# Example dataset
sample_data = [12, 15, 14, 10, 13, 15, 11, 14, 13, 12]

# Calculate 95% confidence interval
ci_lower, ci_upper = confidence_interval(sample_data, confidence=0.95)

# Print results
print(f"95% Confidence Interval for the mean: ({ci_lower:.2f}, {ci_upper:.2f})")
```

```python
#17.Generate data from a normal distribution, then calculate and interpret the confidence interval for its mean.

import numpy as np
from scipy import stats

# Step 1: Generate data from a normal distribution
np.random.seed(42)  # For reproducibility
mu = 100            # True population mean
sigma = 15          # Population standard deviation
n = 50              # Sample size

# Sample data
sample_data = np.random.normal(loc=mu, scale=sigma, size=n)

# Step 2: Calculate the sample mean and standard error
sample_mean = np.mean(sample_data)
sem = stats.sem(sample_data)  # Standard error of the mean

# Step 3: Calculate the 95% confidence interval using the t-distribution
confidence = 0.95
df = n - 1
t_crit = stats.t.ppf((1 + confidence) / 2, df)
margin_of_error = t_crit * sem
ci_lower = sample_mean - margin_of_error
ci_upper = sample_mean + margin_of_error

# Step 4: Display results
print(f"Sample Mean: {sample_mean:.2f}")
print(f"95% Confidence Interval: ({ci_lower:.2f}, {ci_upper:.2f})")
```

```python
#19.Use Python to calculate and interpret the cumulative distribution function (CDF) of a Poisson distribution.

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import poisson

# Parameters
lambda_ = 4  # Average rate (mean)
k_values = np.arange(0, 15)  # Values of k to compute CDF for

# Calculate the CDF
cdf_values = poisson.cdf(k_values, mu=lambda_)

# Display the results
for k, cdf in zip(k_values, cdf_values):
    print(f"P(X ≤ {k}) = {cdf:.4f}")

# Plotting the CDF
plt.figure(figsize=(8, 5))
plt.step(k_values, cdf_values, where='post', label=f'Poisson CDF (λ={lambda_})')
plt.xlabel('k')
plt.ylabel('P(X ≤ k)')
plt.title('CDF of Poisson Distribution')
plt.grid(True)
plt.legend()
plt.show()

from scipy.stats import poisson
cdf_at_6 = poisson.cdf(6, mu=4)
print(f"P(X ≤ 6) = {cdf_at_6:.4f}")
```

```python
#21.Write a Python program to compare the standard deviations of two datasets and visualize the difference.

import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Generate two datasets
np.random.seed(42)  # for reproducibility

data1 = np.random.normal(loc=50, scale=5, size=1000)   # Mean=50, SD=5
data2 = np.random.normal(loc=50, scale=15, size=1000)  # Mean=50, SD=15

# Calculate standard deviations
std1 = np.std(data1)
std2 = np.std(data2)

print(f"Standard Deviation of Dataset 1: {std1:.2f}")
print(f"Standard Deviation of Dataset 2: {std2:.2f}")

# Plotting
plt.figure(figsize=(10, 6))
sns.histplot(data1, color='skyblue', label=f'Dataset 1 (SD={std1:.2f})', kde=True, stat="density")
sns.histplot(data2, color='salmon', label=f'Dataset 2 (SD={std2:.2f})', kde=True, stat="density", alpha=0.6)

plt.title('Comparison of Standard Deviations Between Two Datasets')
plt.xlabel('Value')
plt.ylabel('Density')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
#22.Calculate the range and interquartile range (IQR) of a dataset generated from a normal distribution.

import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Generate a dataset from a normal distribution
np.random.seed(42)
data = np.random.normal(loc=100, scale=15, size=1000)  # Mean=100, SD=15

# Calculate Range
data_range = np.max(data) - np.min(data)

# Calculate IQR
q1 = np.percentile(data, 25)
q3 = np.percentile(data, 75)
iqr = q3 - q1

# Output results
print(f"Range: {data_range:.2f}")
print(f"IQR (Interquartile Range): {iqr:.2f}")

# Optional: Visualize with boxplot
plt.figure(figsize=(8, 4))
sns.boxplot(data, color='lightblue')
plt.title("Boxplot Showing Range and IQR")
plt.xlabel("Value")
plt.grid(True)
plt.show()
```

```python
#23.Implement Z-score normalization on a dataset and visualize its transformation.

import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Generate synthetic data from a normal distribution
np.random.seed(42)
data = np.random.normal(loc=100, scale=20, size=1000)  # Mean=100, SD=20

# Z-score normalization
mean = np.mean(data)
std = np.std(data)
z_scores = (data - mean) / std

# Verify transformation
print(f"Original Mean: {mean:.2f}, Standard Deviation: {std:.2f}")
print(f"Z-score Mean: {np.mean(z_scores):.2f}, Standard Deviation: {np.std(z_scores):.2f}")

# Plot before and after normalization
plt.figure(figsize=(12, 5))

# Original data
plt.subplot(1, 2, 1)
sns.histplot(data, kde=True, color='skyblue')
plt.title("Original Data")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.grid(True)

# Z-score normalized data
plt.subplot(1, 2, 2)
sns.histplot(z_scores, kde=True, color='salmon')
plt.title("Z-Score Normalized Data")
plt.xlabel("Z-score")
plt.ylabel("Frequency")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
#24.Write a Python function to calculate the skewness and kurtosis of a dataset generated from a normal distribution.

import numpy as np
from scipy.stats import skew, kurtosis

def analyze_distribution_stats(data):
    skewness = skew(data)
    kurt = kurtosis(data)  # excess kurtosis by default
    print(f"Skewness: {skewness:.4f}")
    print(f"Excess Kurtosis: {kurt:.4f}")
    return skewness, kurt

# Generate data from normal distribution
np.random.seed(42)
normal_data = np.random.normal(loc=0, scale=1, size=1000)

# Call the function
analyze_distribution_stats(normal_data)
```
